```python
import pandas as pd

lp = pd.read_csv('LOCAL_PEOPLE_GU_2021.csv', encoding='cp949', index_col=False)
lf = pd.read_csv('LONG_FOREIGNER_GU_2021.csv', encoding='cp949', index_col=False)
tf = pd.read_csv('TEMP_FOREIGNER_GU_2021.csv', encoding='cp949', index_col=False)

lp
```

```python
lp['기준일ID'] = lp['기준일ID'].astype('str')
lp['기준일ID']
```

```python
lp['기준일ID'] = pd.to_datetime(lp['기준일ID'])
lp
```

```python
lf['기준일ID'] = lf['기준일ID'].astype('str')
tf['기준일ID'] = tf['기준일ID'].astype('str')

lf['기준일ID'] = pd.to_datetime(lf['기준일ID'])
tf['기준일ID'] = pd.to_datetime(tf['기준일ID'])
```

```python
lp['요일'] = lp['기준일ID'].dt.day_name()
lp
```

```python
lf['요일'] = lf['기준일ID'].dt.day_name()
tf['요일'] = tf['기준일ID'].dt.day_name()
```

```python
lp['월'] = lp['기준일ID'].dt.strftime('%m')
lp
```

```python
lf['월'] = lf['기준일ID'].dt.strftime('%m')
tf['월'] = tf['기준일ID'].dt.strftime('%m')
```

```python
lp.to_csv('LOCAL_PEOPLE_2021.csv', index=False)
lf.to_csv('LONG_FOREIGNER_2021.csv', index=False)
tf.to_csv('TEMPORARY_FOREIGNER_2021.csv', index=False)
```
