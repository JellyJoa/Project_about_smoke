```python
import pandas as pd

people = pd.read_csv('2021전처리.csv', index_col=[0,1])
rate = pd.read_csv('2021흡연율.csv', index_col=0)

people
```

```python
rate
```

```python
smoker = 0

for column, smoke_rate in zip(people, rate['흡연율']):
    smoker += people[column].apply(lambda x: x * smoke_rate / 100)

smoker
```

```python
people['총예상흡연자수'] = smoker
people
```

```python
people = people.round()
people = people.astype('int')
people
```

```python
import os
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

if os.name == 'nt':
    font_family='Malgun Gothic'
else:
    font_family='AppleGothic'
    
sns.set(font=font_family, rc={'axes.unicode_minus':False})
```

```python
people.plot.bar()
plt.title('2021년 자치구별 생활인구의 흡연자수 예측')
plt.xlabel('dd')
plt.ylabel('rr')

plt.show()
```
