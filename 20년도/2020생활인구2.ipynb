```python
import pandas as pd

lp = pd.read_csv('LOCAL_PEOPLE_2020.csv', index_col=[0,1,2])
lf = pd.read_csv('LONG_FOREIGNER_2020.csv', index_col=[0,1,2])
tf = pd.read_csv('TEMPORARY_FOREIGNER_2020.csv', index_col=[0,1,2])

lp
```

```python
lp.drop(['총생활인구수', '요일', '월'], axis=1, inplace=True)
lp
```

```python
lf.drop(['총생활인구수', '요일', '월'], axis=1, inplace=True)
tf.drop('총생활인구수', axis=1, inplace=True)
```

```python
people = pd.concat([lp, lf, tf], axis=1)
people
```

```python
people.drop(['남자0세부터9세생활인구수', '남자10세부터14세생활인구수', '여자0세부터9세생활인구수', '여자10세부터14세생활인구수'], axis=1, inplace=True)
people
```

```python
people.to_csv('2020생활인구.csv')
```
